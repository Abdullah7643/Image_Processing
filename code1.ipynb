## Q1a: Area of image

```python
def area_img(img):
#   converting numpy array into list
    img1=img.tolist()
#   Number of row and coloumns of pixels in image
    print("number of pixels rows in image ",len(img1))
    print("number of pixels columns in image ",len(img1[0]))
#   Number of channel in image
    print("number of colour channels in image ",len(img1[0][0]))
#   Area of image = rows * columns
    print("Spatial Area of image in pixels = ",len(img1)*len(img[0])," pixels")
```

## Q1b,c: Number of white and grey pixels in image

```python
def unique(list1):

    # initialize a null list
    unique_list = []

    # traverse for all elements
    for x in list1:
        # check if exists in unique_list or not
        if x not in unique_list:
            unique_list.append(x)
    # print list
    for x in unique_list:
        print(x)
    return unique_list
```

```python
def flat(list1):
    # initialize a null list
    unique_list = []
    z=[]
    # to make a 2d list
    for x in list1:
        for y in x:
            unique_list.append(y)
    # to make a 1d list
    for x in unique_list:
        z.append(x[0])
    return z
```

```python
# converting the image to 1d list so we can easily get the unique values
x=flat(img1)
len(x)
```

```python
#unique values are taken to find the range of pixel values for each colour
x=unique(x)
```

```python
def color_values(unique_vals):
    white=[]
    grey=[]
    black=[]
    for i in unique_vals:
        if (i>200):
            white.append(i)
        elif (i>100) and (i<200):
            grey.append(i)
        else:
            black.append(i)
    return black,grey,white
```

```python
#creating lists of colours for their values in the image
black,grey,white=color_values(x)
white,grey,black
```

```python
def pixel_counter(img,c1,c2):
    grey=0
    whit=0
    list1=flat(img)
    for x in c1:
        for y in list1:
            if (x==y):
                whit+=1
    for x in c2:
        for y in list1:
            if (y==x):
                grey +=1
    print("Number of white Pixels=",whit)
    print("Number of grey Pixels=",grey)
```

```python
#counting the number of pixels for grey and white colour
pixel_counter(img1,white,grey)
```

## Q3:Thresholding operations in OpenCV

```python
path = r'coins.jpg'
coin = cv2.imread(path)
plt.imshow(cv2.imread(path))
```

```python
#Binary thres
_,t1 = cv2.threshold(coin,100,255,cv2.THRESH_BINARY)
#Binary_INV thres
_,t2 = cv2.threshold(coin,100,255,cv2.THRESH_BINARY_INV)
#truncate
_,t3 = cv2.threshold(coin,100,255,cv2.THRESH_TRUNC)
#threshold toZero
_,t4 = cv2.threshold(coin,100,255,cv2.THRESH_TOZERO)
#Threshold toZero_INV
_,t5 = cv2.threshold(coin,100,255,cv2.THRESH_TOZERO_INV)
```

```python
thresholds = ['Original Image','Thresholding BINARY','Thresholding BINARY_INVER','Thresholding TRUNCATE','Thresholding TOZERO','Thresholding TOZERO_INVER']
imgs = [coin, t1, t2, t3, t4, t5]
for i in range(6):
    plt.figure(figsize=(5,5))
    plt.imshow(imgs[i])
    plt.title(thresholds[i])
plt.show()
```

## Q5:Segmentation of Brain using CTSCAN

```python
path = r'ctscan.tif'
ctscan = cv2.imread(path)
seg_ct=cv2.imread(path)
plt.imshow(cv2.imread(path))
```

```python
_, seg_img = cv2.threshold(cv2.imread(path), 250, 255, cv2.THRESH_BINARY)
row,col=0,0
for z in seg_img:
    col=0
    for y in z:
        if (y[0]>200 and row>160):
            seg_ct[row][col]=[255,0,0]
        col+=1
    row+=1
```

```python
ctscans = ['Original Image','Segmentation Result']
images = [ctscan, seg_ct]
for i in range(2):
    plt.figure(figsize=(10,10))
    plt.subplot(1,2,i+1),plt.imshow(images[i])
    plt.title(ctscans[i])
plt.show()
```
